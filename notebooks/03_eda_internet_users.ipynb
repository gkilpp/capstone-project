```python
import pandas as pd
```

```python
df_internet = pd.read_csv("../data/raw/internet_users_dataset.csv", skiprows=4)
```

```python
df_turkey = df_internet[df_internet["Country Name"].str.contains("Tur", case=False)]
df_turkey.head()
```

```python
countries =  ["United States", "Germany", "Brazil", "Turkiye", "Thailand", "Viet Nam", "Canada", "United Kingdom", "Japan", "India"]

df_internet = df_internet[df_internet['Country Name'].isin(countries)]

df_internet["Country Name"] = df_internet["Country Name"].replace({
    "Viet Nam": "Vietnam", "Turkiye": "Turkey"
})
```

```python
df_internet['Country Name'].unique()
```

```python
df_internet.head()
```

```python
df_internet['internet_users'] = df_internet[['2016', '2017']].mean(axis=1)
```

```python
df_internet = df_internet[['Country Name', 'internet_users', "Country Code"]]
```

```python
df_internet.columns = ['country', 'internet_users', 'country_code']
```

```python
df_internet['country'] = df_internet['country'].str.strip()
```

```python
df_internet = df_internet.reset_index(drop=True)
df_internet.head(10)
```

```python
df_internet.rename(columns={'internet_users': 'avg_internet_usage_pct'}, inplace=True)
```

```python
df_internet['avg_internet_usage_pct'] = df_internet['avg_internet_usage_pct'].round(2)
```
