```python

usa_clean = pd.read_csv("../data/processed/usa_clean.csv")
germany_clean = pd.read_csv("../data/processed/germany_clean.csv")
brazil_clean = pd.read_csv("../data/processed/brazil_clean.csv")
turkey_clean = pd.read_csv("../data/processed/turkey_clean.csv")
thailand_clean = pd.read_csv("../data/processed/thailand_clean.csv")
uk_clean = pd.read_csv("../data/processed/uk_clean.csv")
canada_clean = pd.read_csv("../data/processed/canada_clean.csv")
india_clean = pd.read_csv("../data/processed/india_clean.csv")
vietnam_clean = pd.read_csv("../data/processed/vietnam_clean.csv")
japan_clean = pd.read_csv("../data/processed/japan_clean.csv")

```

```python
dfs = [usa_clean, germany_clean, brazil_clean, turkey_clean, thailand_clean,
       uk_clean, canada_clean, india_clean, vietnam_clean, japan_clean]

df_all = pd.concat(dfs, ignore_index=True)
```

```python
df_all.head()
```

```python
df_all['country'].unique()
```

```python
df_all['date'] = pd.to_datetime(df_all['date'])
```

```python
df_all['date'] = df_all['date'].dt.tz_localize(None)
```

```python
df_all
```

```python
df_all.isnull().sum()
```

```python
df_all = df_all.rename(columns={
    'temperature_2m_mean': 'temp_mean',
    'temperature_2m_max': 'temp_max',
    'temperature_2m_min': 'temp_min',
    'wind_speed_10m_max': 'wind_speed',
    'precipitation_sum': 'precipitation',
    'rain_sum': 'rain',
    'snowfall_sum': 'snow',
    'precipitation_hours': 'precip_hours',
    'wind_gusts_10m_max': 'wind_gusts'
})
```

```python
df_all.head()
```

```python
print(df_all.duplicated().sum())
```

```python
df_all = df_all.reset_index(drop=True)
```

```python
df_all.head(10)
```

```python
df_all.dtypes
```

```python
df_all[['temp_mean', 'temp_max', 'temp_min', 'wind_speed', 'precipitation', 'rain', 'wind_gusts']] = df_all[['temp_mean', 'temp_max', 'temp_min', 'wind_speed', 'precipitation', 'rain', 'wind_gusts']].round(1)
```

```python
df_all['sunshine_hours'] = df_all['sunshine_duration'] / 3600
df_all['daylight_hours'] = df_all['daylight_duration'] / 3600
```

```python
df_all = df_all.drop(['sunshine_duration', 'daylight_duration'], axis=1)
```

```python
df_all[['sunshine_hours', 'daylight_hours']] = df_all[['sunshine_hours', 'daylight_hours']].round(1)
```

```python
df_all = df_all.drop(['sunset', 'sunrise'], axis=1)
```
