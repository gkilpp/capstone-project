```python
df_weather_turkey = df_weather[(df_weather['country']=='Turkey') & (df_weather['precipitation'] > 0)]
```

```python
print(df_weather_turkey)
```

```python
df_weather = df_weather[df_weather['country'] != "Turkey"]
```

```python
df_weather.head()
```

```python
df_weather['country'].unique()
```

```python
df_weather.isnull().sum()
```

```python
# Nan values were replaced with null values because the analysis is guided through extreme weather events
df_weather['precipitation'] = df_weather['precipitation'].fillna(0)
```

```python
df_weather['temp_avg'] = df_weather['temp_avg'].fillna(df_weather['temp_avg'].mean())
```

```python
df_weather['temp_min'] = df_weather['temp_min'].fillna(df_weather['temp_avg'])
df_weather['temp_max'] = df_weather['temp_max'].fillna(df_weather['temp_avg'])
```

```python
df_weather.to_csv("../data/processed/weather_clean.csv", index=False)
```
