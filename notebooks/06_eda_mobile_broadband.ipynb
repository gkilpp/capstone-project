```python
import pandas as pd
```

```python

df_mobile_broadband = pd.read_csv("../data/raw/active_mobile_broadband_subscriptions.csv")
```

```python
df_mobile_broadband.head(10)
```

```python
df_mobile_broadband["Economy"] = df_mobile_broadband["Economy"].replace({
    "Viet Nam": "Vietnam", "Türkiye": "Turkey"})
```

```python
df_mobile_broadband.head(10)
```

```python
df_mobile_broadband['mobile_broadband_subscriptions_per_100_people'] = df_mobile_broadband[['2016', '2017']].mean(axis=1)
```

```python
df_mobile_broadband = df_mobile_broadband[['Economy', 'mobile_broadband_subscriptions_per_100_people']]
df_mobile_broadband.columns = ['country', 'mobile_broadband_subscriptions_per_100_people']
```

```python
df_mobile_broadband.head(10)
```

```python
df_mobile_broadband['mobile_broadband_subscriptions_per_100_people'] = (
    df_mobile_broadband['mobile_broadband_subscriptions_per_100_people'].round(2))
```

```python
df_mobile_broadband.head()
```

```python
df_mobile_broadband.to_csv("../data/processed/mobile_broadband_clean.csv", index=False)
```
