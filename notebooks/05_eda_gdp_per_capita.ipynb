```python
import pandas as pd
```

```python
df_gdp = pd.read_csv("../data/raw/gdp_per_capita.csv", skiprows=4)
```

```python
countries = ["United States", "Germany", "Brazil", "Turkiye", "Thailand", "Viet Nam", "Canada", "United Kingdom", "Japan", "India"]

df_gdp = df_gdp[df_gdp['Country Name'].isin(countries)]

df_gdp["Country Name"] = df_gdp["Country Name"].replace({
    "Viet Nam": "Vietnam", "Turkiye": "Turkey"})
```

```python
df_gdp['gdp_per_capita'] = df_gdp[['2016', '2017']].mean(axis=1)
```

```python
df_gdp = df_gdp[['Country Name', 'gdp_per_capita', "Country Code"]]
```

```python
df_gdp.columns = ['country', 'gdp_per_capita', "country_code"]
```

```python
df_gdp.dtypes
```

```python
df_gdp = df_gdp.reset_index(drop=True)
```

```python
df_gdp.to_csv("../data/processed/gdp_clean.csv", index=False)
```
