"Fixed broadband subscriptions (per 100 people)"

"Fixed broadband subscriptions refers to fixed subscriptions to high-speed access to the public Internet (a TCP/IP connection), at downstream speeds equal to, or greater than, 256 kbit/s. This includes cable modem, DSL, fiber-to-the-home/building, other fixed (wired)-broadband subscriptions, satellite broadband and terrestrial fixed wireless broadband. This total is measured irrespective of the method of payment. It excludes subscriptions that have access to data communications (including the Internet) via mobile-cellular networks. It should include fixed WiMAX and any other fixed wireless technologies. It includes both residential subscriptions and subscriptions for organizations.","World Telecommunication/ICT Indicators Database, International Telecommunication Union (ITU)",



```python
import pandas as pd
```

```python
df_fixed_broadband = pd.read_csv("../data/raw/fixed_broadband_subscriptions.csv", skiprows=4)
```

```python
countries =  ["United States", "Germany", "Brazil", "Turkiye", "Thailand", "Viet Nam", "Canada", "United Kingdom", "Japan", "India"]

df_fixed_broadband = df_fixed_broadband[df_fixed_broadband['Country Name'].isin(countries)]

df_fixed_broadband["Country Name"] = df_fixed_broadband["Country Name"].replace({
    "Viet Nam": "Vietnam", "Turkiye": "Turkey"})
```

```python
df_fixed_broadband['broadband'] = df_fixed_broadband[['2016', '2017']].mean(axis=1)

df_fixed_broadband = df_fixed_broadband[['Country Name', 'broadband', "Country Code"]]
df_fixed_broadband.columns = ['country', 'broadband', 'country_code']
```

```python
df_fixed_broadband.rename(columns={
    'broadband': 'fixed_broadband_subscriptions_per_100_people'
}, inplace=True)
```

```python
df_fixed_broadband = df_fixed_broadband.reset_index(drop=True)
df_fixed_broadband.head(10)
```

```python
df_fixed_broadband['fixed_broadband_subscriptions_per_100_people'] = (
    df_fixed_broadband['fixed_broadband_subscriptions_per_100_people'].round(2)
)
```
